# P105 - Product Cannibalization & Demand Shift Analysis

```python
# PHASE 1: Data Loading & Understanding

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. LOAD DATASET

print("=" * 60)
print("  P105 | PHASE 1: DATA LOADING & UNDERSTANDING")
print("=" * 60)

df = pd.read_csv("D:\\Portfolio\\Creations\\Projects\\P105 Product Cannibalization & Demand Shift Analysis\\Data\\advanced_cannibalization_dataset.csv")

print(f"\n✅ Dataset loaded successfully.")
print(f"   Rows    : {df.shape[0]:,}")
print(f"   Columns : {df.shape[1]}")
```

```python
# 2. BASIC STRUCTURE

print("\n" + "=" * 60)
print("  SECTION 2 | DATASET STRUCTURE")
print("=" * 60)

print("\n[Column Names & Data Types]")
print(df.dtypes.to_string())

print("\n[First 5 Rows]")
print(df.head().to_string())

print("\n[Last 5 Rows]")
print(df.tail().to_string())
```

```python
# 3. MISSING VALUES

print("\n" + "=" * 60)
print("  SECTION 3 | MISSING VALUE ANALYSIS")
print("=" * 60)

null_counts = df.isnull().sum()
null_pct    = (null_counts / len(df) * 100).round(2)
null_report = pd.DataFrame({'Missing Count': null_counts, 'Missing %': null_pct})
print(null_report.to_string())

if null_counts.sum() == 0:
    print("\n✅ No missing values found. Dataset is complete.")
else:
    print(f"\n⚠️  Total missing values: {null_counts.sum()}")
```

```python
# 4. DUPLICATE RECORDS

print("\n" + "=" * 60)
print("  SECTION 4 | DUPLICATE RECORDS")
print("=" * 60)

dupes = df.duplicated().sum()
print(f"   Duplicate rows: {dupes}")
if dupes == 0:
    print("✅ No duplicate records found.")
else:
    print(f"⚠️  {dupes} duplicates detected.")
```

```python
# 5. DESCRIPTIVE STATISTICS

print("\n" + "=" * 60)
print("  SECTION 5 | DESCRIPTIVE STATISTICS (Numerical Columns)")
print("=" * 60)

num_cols = ['Price', 'Rating', 'Marketing_Spend', 'Stock_Available', 'Sales', 'Revenue']
print(df[num_cols].describe().round(2).to_string())
```

```python
# 6. CATEGORICAL VARIABLE OVERVIEW

print("\n" + "=" * 60)
print("  SECTION 6 | CATEGORICAL VARIABLES")
print("=" * 60)

# Products
print(f"\n[Unique Products] — {df['Product_ID'].nunique()} total")
print(sorted(df['Product_ID'].unique()))

# Categories
print(f"\n[Category Distribution]")
cat_dist = df['Category'].value_counts().reset_index()
cat_dist.columns = ['Category', 'Record Count']
cat_dist['%'] = (cat_dist['Record Count'] / len(df) * 100).round(1)
print(cat_dist.to_string(index=False))

# Product Groups
print(f"\n[Product Group → Products Mapping]")
pg_map = df.groupby('Product_Group')['Product_ID'].apply(lambda x: sorted(x.unique())).reset_index()
pg_map.columns = ['Product_Group', 'Products']
print(pg_map.to_string(index=False))

# Regions
print(f"\n[Region Distribution]")
region_dist = df['Region'].value_counts().reset_index()
region_dist.columns = ['Region', 'Record Count']
region_dist['%'] = (region_dist['Record Count'] / len(df) * 100).round(1)
print(region_dist.to_string(index=False))

# Launch Flag
print(f"\n[Launch Flag Distribution]")
launch_dist = df['Launch_Flag'].value_counts().reset_index()
launch_dist.columns = ['Launch_Flag', 'Count']
launch_dist['Meaning'] = launch_dist['Launch_Flag'].map({0: 'Normal sales period', 1: 'Product launch period'})
print(launch_dist.to_string(index=False))
```

```python
# 7. DATE RANGE

print("\n" + "=" * 60)
print("  SECTION 7 | DATE RANGE")
print("=" * 60)

df['Date'] = pd.to_datetime(df['Date'])
print(f"   Earliest date : {df['Date'].min().date()}")
print(f"   Latest date   : {df['Date'].max().date()}")
print(f"   Date span     : {(df['Date'].max() - df['Date'].min()).days} days (~2 years)")

monthly = df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum()
print(f"\n   Monthly sales range: {monthly.min():,} – {monthly.max():,} units")
```

```python
# 8. LAUNCHED PRODUCTS DETECTION

print("\n" + "=" * 60)
print("  SECTION 8 | LAUNCHED PRODUCTS IDENTIFICATION")
print("=" * 60)

launched = df[df['Launch_Flag'] == 1].groupby('Product_ID').agg(
    Launch_Start=('Date', 'min'),
    Launch_End=('Date', 'max'),
    Launch_Records=('Launch_Flag', 'count')
).reset_index()

print(f"\n   Products with Launch_Flag=1: {launched['Product_ID'].nunique()}")
print(launched.to_string(index=False))
print("\n   ➤ Launch window: 2024-06-01 onwards")
print("   ➤ Launched products: P1, P2, P3 (Group G1) | P4, P5 (Group G2)")
```

```python
# 9. STOCK AVAILABILITY

print("\n" + "=" * 60)
print("  SECTION 9 | STOCK AVAILABILITY")
print("=" * 60)

stock_dist = df['Stock_Available'].value_counts().reset_index()
stock_dist.columns = ['Stock_Available', 'Count']
stock_dist['Meaning'] = stock_dist['Stock_Available'].map({1: 'In Stock', 0: 'Out of Stock'})
stock_dist['%'] = (stock_dist['Count'] / len(df) * 100).round(1)
print(stock_dist.to_string(index=False))

out_of_stock_products = df[df['Stock_Available'] == 0].groupby('Product_ID')['Sales'].count()
print(f"\n   Out-of-stock records per product:")
print(out_of_stock_products.sort_values(ascending=False).to_string())

```

```python
# 10. SUMMARY CARD

print("\n" + "=" * 60)
print("  PHASE 1 SUMMARY")
print("=" * 60)
print(f"  Total Records      : {len(df):,}")
print(f"  Total Columns      : {df.shape[1]}")
print(f"  Products           : {df['Product_ID'].nunique()}")
print(f"  Product Groups     : {df['Product_Group'].nunique()}")
print(f"  Categories         : {df['Category'].nunique()}")
print(f"  Regions            : {df['Region'].nunique()}")
print(f"  Date Range         : {df['Date'].min().date()} to {df['Date'].max().date()}")
print(f"  Missing Values     : 0")
print(f"  Duplicates         : 0")
print(f"  Launched Products  : P1, P2, P3, P4, P5 (from 2024-06-01)")
print(f"  Out-of-Stock Rows  : {(df['Stock_Available']==0).sum():,}")
print(f"  Total Revenue      : ₹{df['Revenue'].sum():,.0f}")
print(f"  Total Units Sold   : {df['Sales'].sum():,}")
print("\n✅ Phase 1 Complete.\n")
```
